### ¿Cuántas personas lo usan cada día, mes y semana?  
Para responder a esta pregunta usaremos los datos de la tabla de visitas. Haremos agrupaciones por día, semana y mes.

```python
# mostramos parte del dataframe que contiene los registros de las visitas
visitas_datos.head(1)
```

```python
"""Definamos una función para calcular las métricas de uso"""
def metrics_use(data):
    """data= el dataframe que contiene los datos de las visitas"""
    
    # cantidad promedio diaria de personas que usan la aplicación, 'DAU'
    dau=math.ceil(data.groupby('date').agg({'uid':'nunique'}).mean())
    print(f'{dau=}',end='\n')
    
    # cantidad promedio semanal de personas que usan la aplicación, 'WAU'
    wau=math.ceil(data.groupby(['month','week']).agg({'uid':'nunique'}).mean())
    print(f'{wau=}',end='\n')
    
    # cantidad promedio mensual de personas que usan la aplicación, 'MAU'
    mau=math.ceil(data.groupby('month').agg({'uid':'nunique'}).mean())
    print(f'{mau=}')
    return dau,wau,mau
```

```python
# calculamos las metricas de uso diario en general
dau,wau,mau=metrics_use(data=visitas_datos)
```

```python
# Creamos un gráfico tipo 'funnel' para mostrar la cantidad de personas que usan la aplicación
df = pd.DataFrame(dict(
    periodo = ['Mensuales', 'Semanales', 'Diarios'],
    cantidad = [mau, wau, dau]))

fig, ax = plt.subplots()
ax.barh(df['periodo'], df['cantidad'], color=['blue', '#FFA07A', '#FFA00B'])
ax.invert_yaxis() # Los datos van de top a bottom
for i, val in enumerate(df['cantidad']):
        ax.text(val, i, str(val))
ax.set_xlabel('Número de usuarios')
ax.set_title('Cantidad promedio de usuarios que usan la aplicación')
plt.xlim(0,df['cantidad'].max()*1.2)
plt.show()
```

---
Tenemos entonces que **en promedio la aplicación es usada:**

- **diariamente por 908 personas (DAU)**
- **semanalmente por 4,913 personas (WAU)**
- **mensualmente por 23,229 personas (MAU)**

```python
# calculamos las metricas de uso diario para los aparatos 'touch'
dau,wau,mau=metrics_use(data=visitas_datos.query('device=="touch"'))
```

```python
# Creamos un gráfico tipo 'funnel' para mostrar la cantidad de personas que usan la aplicación
device="'touch'"
df = pd.DataFrame(dict(
    periodo = ['Mensuales', 'Semanales', 'Diarios'],
    cantidad = [mau, wau, dau]))

fig, ax = plt.subplots()
ax.barh(df['periodo'], df['cantidad'], color=['blue', '#FFA07A', '#FFA00B'])
ax.invert_yaxis() # Los datos van de top a bottom
for i, val in enumerate(df['cantidad']):
        ax.text(val, i, str(val))
ax.set_xlabel('Número de usuarios')
ax.set_title("Cantidad promedio de usuarios que usan la aplicación usando aparatos "+device)
plt.xlim(0,df['cantidad'].max()*1.2)
plt.show()
```

---
Tenemos entonces que **en promedio la aplicación es usada en los aparatos `'touch'`:**

- **diariamente por 253 personas (DAU)**
- **semanalmente por 1,404 personas (WAU)**
- **mensualmente por 6,826 personas (MAU)**

```python
# calculamos las metricas de uso diario para los aparatos 'desktop'
dau,wau,mau=metrics_use(data=visitas_datos.query('device=="desktop"'))
```

```python
# Creamos un gráfico tipo 'funnel' para mostrar la cantidad de personas que usan la aplicación
device="'desktop'"
df = pd.DataFrame(dict(
    periodo = ['Mensuales', 'Semanales', 'Diarios'],
    cantidad = [mau, wau, dau]))

fig, ax = plt.subplots()
ax.barh(df['periodo'], df['cantidad'], color=['blue', '#FFA07A', '#FFA00B'])
ax.invert_yaxis() # Los datos van de top a bottom
for i, val in enumerate(df['cantidad']):
        ax.text(val, i, str(val))
ax.set_xlabel('Número de usuarios')
ax.set_title("Cantidad promedio de usuarios que usan la aplicación usando aparatos "+device)
plt.xlim(0,df['cantidad'].max()*1.2)
plt.show()
```

### ¿Cuántas sesiones hay por día, mes y semana?  
Para responder a esta pregunta usaremos los datos de la tabla de visitas. Haremos agrupaciones por día, semana y mes.

```python
"""Definamos una función para calcular las métricas de las sesiones"""
def metrics_session(data):
    """data= el dataframe que contiene los datos de las visitas
        session_day=sesiones diarias
        session_week=sesiones semanales
        session_month=sesiones mensuales
    """
    
    # cantidad promedio diaria de sesiones en la aplicación
    session_day=math.ceil(data.groupby('date').agg({'uid':'count'}).mean())
    print(f'{session_day=}',end='\n')
    
    # cantidad promedio semanal de sesiones en la aplicación
    session_week=math.ceil(data.groupby('week').agg({'uid':'count'}).mean())
    print(f'{session_week=}',end='\n')
    
    # cantidad promedio mensual de sesiones en la aplicación
    session_month=math.ceil(data.groupby('month').agg({'uid':'count'}).mean())
    print(f'{session_month=}')
    return session_day,session_week,session_month
```

```python
# calculamos las metricas de las sesiones
session_day,session_week,session_month=metrics_session(data=visitas_datos)
```

```python
# Creamos un gráfico tipo 'funnel' para mostrar la cantidad de sesiones que se inician en la aplicación

df = pd.DataFrame(dict(
    periodo = ['Mensuales', 'Semanales', 'Diarios'],
    cantidad = [session_month, session_week, session_day]))

fig, ax = plt.subplots()
ax.barh(df['periodo'], df['cantidad'], color=['blue', '#FFA07A', '#FFA00B'])
ax.invert_yaxis() # Los datos van de top a bottom
for i, val in enumerate(df['cantidad']):
        ax.text(val, i, str(val))
ax.set_xlabel('Número de sesiones')
ax.set_title("Cantidad promedio de sesiones que se inician en la aplicación")
plt.xlim(0,df['cantidad'].max()*1.2)
plt.show()
```

---
Tenemos entonces que **en promedio en la aplicación se inician:**

- **diariamente 988 sesiones**
- **semanalmente 6,782 sesiones**
- **mensualmente 29,950 sesiones**

```python
# calculamos las metricas de las sesiones para los aparatos 'touch'
metrics_session(data=visitas_datos.query('device=="touch"'))
```

---

Podemos observar que el 75% de los pedidos se realizan en los primeros cuatro días después que un usuario se registra. **Vamos a crear seis cohortes:**
- **'Conversion 0d',** para aquellos usuarios cuya primer compra la hicieron el mismo día de su registro.
- **'Conversion 1d',** para aquellos usuarios cuya primer compra la hicieron un día posterior a su registro.
- **'Conversion 2d',** para aquellos usuarios cuya primer compra la hicieron dos días posteriores a su registro.
- **'Conversion 3d',**  para aquellos usuarios cuya primer compra la hicieron tres días posteriores a su registro.
- **'Conversion 4d',**  para aquellos usuarios cuya primer compra la hicieron cuatro días posteriores a su registro.
- **'Conversion 4d+',**  para aquellos usuarios cuya primer compra la hicieron después del cuarto día posterior a su registro.

```python
"""Creamos una función para asignar los cohortes en base a los días que pasa entre el registro y 
la compra de un usuario"""
def cohorte_user(row):
    dias=row['days_to_order']
    
    if (dias <=0):
        return 'Conversion 0d'
    elif (0<dias<=1):
        return 'Conversion 1d'
    elif (1<dias<=2):
        return 'Conversion 2d'
    elif (2<dias<=3):
        return 'Conversion 3d'
    elif (3<dias<=4):
        return 'Conversion 4d'
    else:
        return 'Conversion 4d+'
```

```python
# creamos la columna de cohortes
registers_datos['cohorte_conversion']=registers_datos.apply(cohorte_user,axis=1)
```

```python
# comparemos las cohortes creadas en base a su conversión
n=registers_datos.shape[0]
(registers_datos.groupby('cohorte_conversion').agg(percent=('days_to_order','count'))\
                .sort_values(by='percent',ascending=False)/n).style.format('{:.1%}')
```

### ¿Cuánto dinero traen? (LTV)  
Para estudiar el **LTV (LifeTime Value)** usaremos el dataframe **'registers_datos'** ya que este contiene: los datos de los pedidos, la fecha de compra, el mes de compra y la fecha de registro a la plataforma.

Recordemos que **LTV** es la cantidad total de dinero que un cliente aporta a la empresa en promedio al realizar compras.

```python
# obtenemos una vista del dataframe
registers_datos.sample()
```

Para estudiar el **LTV** vamos a crear cohortes mensuales.

```python
# obtengamos el tamaño de las cohortes mensuales en base a la fecha de su primera compra
cohorte_sizes=registers_datos.groupby('first_order_month').agg(new_buyers=('uid','nunique'))
cohorte_sizes.head()
```

```python
# obtengamos el total de ventas por mes para cada cohorte mensual
cohorte_revenue=registers_datos.groupby(['first_order_month','order_month'])\
                                .agg(revenue=('revenue','sum')).reset_index()
print(f'El tamaño del dataframe creado es:',cohorte_revenue.shape)
cohorte_revenue.head()
```

```python
# agregamos ahora el tamaño de los cohortes con su ventas totales por mes
report_pedidos=pd.merge(cohorte_revenue,cohorte_sizes,on='first_order_month')
print(f'El tamaño del dataframe creado es:',report_pedidos.shape)
report_pedidos.head()
```

```python
"""consideremos una tasa de rentabilidad del 30%, 
si consideramos que 'revenue' ya incluye la tasa de ganancia ponemos margin_rate=1 """
margin_rate=1

# obtengamos el beneficio bruto de cada cohorte
report_pedidos['benefit_brute']=report_pedidos['revenue']*margin_rate
```

```python
# obtengamos la edad de cada cohorte en meses
report_pedidos['age']=(report_pedidos['order_month']-report_pedidos['first_order_month'])/np.timedelta64(1,'M')
report_pedidos['age']=report_pedidos['age'].round().astype('int')
report_pedidos.sample()
```

```python
# obtenemos LTV
report_pedidos['ltv']=report_pedidos['benefit_brute']/report_pedidos['new_buyers']
report_pedidos.head()
```

```python
# obtengamos el LTV de cada cohorte por medio de una tabla
output=report_pedidos.pivot_table(index='first_order_month',columns='age',values='ltv',aggfunc='sum').cumsum(axis=1)
output.fillna('')

output.index=output.index.astype(str)
plt.figure(figsize=(10,12))
plt.title('LTV')
ax=sns.heatmap(output,annot=True,fmt='.2f',linecolor='gray',linewidths=.3,\
               cmap='coolwarm',square=True,cbar_kws={'pad': .02,'shrink':0.8},annot_kws={'size': 'small'}) 
ax.xaxis.tick_top()
plt.show()
```

### ¿Cuán rentables eran las inversiones? (ROMI)
Para responder a esta pregunta necesitamos información de las ventas y los costos de marketing mensuales por cada tipo de fuente.

```python
# los gastos mensuales de marketing los obtendremos usando el set de datos'costs_marketing'
costs_marketing.sample()
```

```python
# usaremos 'users' que contiene los datos de las fuentes para los usuarios para unirlo con la tabla de compras
pedidos_source=pedidos_datos.merge(users,on='uid')
```

```python
# echamos un vistazo a los datos
pedidos_source.sample()
```

```python
# obtengamos la fecha de la primera compra
first_order_dt=pedidos_source.groupby('uid').agg(first_order_dt=('buy_date','min'))
```

```python
# unimos la  fecha de la primera compra al dataframe 'pedidos_source'
pedidos_source=pedidos_source.merge(first_order_dt,on='uid')
```

```python
# obtenemos las ventas para cada tipo de fuente segun la fecha de compra
revenue_source= pedidos_source.groupby(['source_id', 'first_order_dt']).agg(revenue=('revenue','sum')).reset_index()
```

Vamos a unir ahora el dataframe **'revenue_source'** que contiene las ventas por fuente para cada fecha de venta con el dataframe **'costs_marketing'** que contiene los costos de marketing.

```python
# revisamos el tamaño de 'costs_marketing' y las caracteristicas de la columna 'first_order_dt'
costs_marketing['first_order_dt'].describe(datetime_is_numeric=True)
```

```python
# revisamos el tamaño de 'revenue_source' y las caracteristicas de la columna 'first_order_dt'
revenue_source['first_order_dt'].describe(datetime_is_numeric=True)
```

```python
# unimos las ventas por mes con los datos por marketing 
costs_revenue_source=costs_marketing.merge(revenue_source,on=['source_id','first_order_dt'])
```

```python
# revisamos el tamaño del dataframe creado y las caracteristicas de la columna 'first_order_dt'
costs_revenue_source['first_order_dt'].describe(datetime_is_numeric=True)
```

Podemos ver que tenemos una perdida de dos registros. Por el momento vamos a ignorar eso y proceder a calcular el 'ROMI' por cada tipo de fuente empleando la siguiente fórmula:
- ROMI=ventas totales / gastos marketing

```python
# agrupamos por fuente y obtenemos las ventas totales y los costos de marketing
source_grouped=costs_revenue_source.groupby('source_id').agg(sum_revenue=('revenue','sum'),sum_costs=('costs','sum'))
```

```python
# calculamos el 'romi' por fuente según la fórmula anterior
source_grouped['romi']=source_grouped['sum_revenue']/source_grouped['sum_costs']
```

```python
# mostramos el 'romi' por fuente
source_grouped
```

```python
# creamos un heatmap para observar el 'romi' por fuente

plt.figure(figsize=(6,8))
plt.title('ROMI por tipo de fuente')
ax=sns.heatmap(source_grouped[['romi']],annot=True,fmt='.2f',linecolor='gray',linewidths=.3,\
               cmap='coolwarm',center=0.9,square=True,cbar_kws={'pad': .04,'shrink':0.9},annot_kws={'size': 'medium'})
plt.xticks(rotation=45)
plt.show()
```

Hasta el momento **las inversiones en marketing no han sido rentables para los tipos de fuentes 3, 4 y 10.** Para los tipos de **fuentes 1, 2, 5 y 9, las inversiones en marketing ya son rentables. Siendo la fuente tipo 1 la de mayor rentabilidad.**

# Anexo  
Cálculo de 'ROMI' promedio por fuente y cohorte mensual

```python
month_revenue = orders.groupby(['uid', 'order_month']).agg({'revenue': 'sum'}).reset_index()
buyers = buyers.merge(month_revenue,on = 'uid')

for source, df in buyers.groupby('source_id'):
    if source !=7:

        cohort_sizes_t = (
            df.groupby(['source_id', 'first_order_month'])
            .agg({'uid': 'nunique'})
            .reset_index()
        )
        cohort_sizes_t.columns = ['source_id', 'first_order_month', 'n_buyers']

        cohorts_revenue_t = (
            df.groupby(['source_id', 'first_order_month', 'order_month'])
            .agg({'revenue': 'sum'})
            .reset_index()
        )
        cohorts_revenue_t.columns = ['source_id', 'first_order_month', 'order_month', 'gp']

        report_romi = pd.merge(cohort_sizes_t, cohorts_revenue_t, on=['source_id', 'first_order_month'])
        report_romi['age'] = (
            report_romi['order_month'] - report_romi['first_order_month']
        ) / np.timedelta64(1, 'M')
        report_romi['age'] = report_romi['age'].round().astype('int')
        report_romi['ltv'] = report_romi['gp'] / report_romi['n_buyers']

        report_romi = pd.merge(report_romi,costs_marketing.groupby(['source_id', 'month_marketing']).agg({'costs': 'sum'}).reset_index(),
                       left_on=['source_id', 'first_order_month'], right_on=['source_id', 'month_marketing'])
        report_romi = report_romi.drop(['month_marketing'], axis = 1) 
        report_romi['cac'] = report_romi['costs'] / report_romi['n_buyers']
        report_romi['romi'] = report_romi['ltv'] / report_romi['cac']


        romi = report_romi\
                .pivot_table(index='first_order_month', columns='age', values='romi', aggfunc='mean')\
                .cumsum(axis=1)
        romi.round(2).fillna('')

        romi.index=romi.index.astype(str)
        sns.heatmap(romi, annot=True, center=0.9,fmt='.2f', linewidths=1, linecolor='grey',cmap='coolwarm',cbar_kws= {'orientation': 'horizontal'} 
            ).set(title ='ROMI by source {}'.format(source))
        plt.show()
        
```
